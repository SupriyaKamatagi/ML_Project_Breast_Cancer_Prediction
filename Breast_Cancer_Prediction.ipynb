
### Breast Cancer Prediction (Morphology)

```python
import pandas as pd
df = pd.read_csv("Breast_cancer.csv")

# The diagnosis is whether the cancer is malignant (1) or benign (0)
df.head()
```

```python
dummy_variables = ['diagnosis'
]
```

```python
df = pd.get_dummies(data = df, prefix=dummy_variables, columns = dummy_variables)
```

```python
df.columns
```

```python
 df['diagnosis_M'] = df['diagnosis_M'].astype(int)
```

```python
df.drop(['diagnosis_B'], axis=1, inplace=True)
```

```python
df.drop(['Unnamed: 32'], axis=1, inplace=True)
```

```python
df = df.rename(columns = {'diagnosis_M': 'diagnosis'})
```

```python
df.columns
```

```python
dict(df.isna().any())
```

```python
df.isna().sum()
```

```python
cancer_type = df['diagnosis'].value_counts()
cancer_type
```

```python
diagnosis = df.groupby('diagnosis').count()['id']
diagnosis.plot(xlabel='diagnosis', ylabel='Frequency', kind='bar')
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=1, test_size=0.3)
```

```python
print(f'train samples: {len(X_train)}')
print(f'testing samples: {len(X_val)}')
print(f'target proportion: {sum(y_train)/len(y_val):.4f}')
```

```python
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()

X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_val)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train_std, y_train)
log_reg.score(X_test_std, y_val)
```

```python
n_estimators_list = [10,20,50,70,100,500]
```

```python
import matplotlib.pyplot as plt
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
RANDOM_STATE = 44
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
accuracy_list_train = []
accuracy_list_val = []
for n_estimators in n_estimators_list:
    # You can fit the model at the same time you define it, because the fit function returns the fitted estimator.
    model = RandomForestClassifier(n_estimators = n_estimators,
                                   random_state = RANDOM_STATE).fit(X_train,y_train)
    predictions_train = model.predict(X_train) ## The predicted values for the train dataset
    predictions_val = model.predict(X_val) ## The predicted values for the test dataset
    accuracy_train = accuracy_score(predictions_train,y_train)
    accuracy_val = accuracy_score(predictions_val,y_val)
    accuracy_list_train.append(accuracy_train)
    accuracy_list_val.append(accuracy_val)

plt.title('Train x Validation metrics')
plt.xlabel('n_estimators')
plt.ylabel('accuracy')
plt.xticks(ticks = range(len(n_estimators_list )),labels=n_estimators_list)
plt.plot(accuracy_list_train)
plt.plot(accuracy_list_val)
plt.legend(['Train','Validation'])
```

```python
n = int(len(X_train)*0.8)
```

```python
X_train_fit, X_train_eval, y_train_fit, y_train_eval = X_train[:n], X_train[n:], y_train[:n], y_train[n:]
```

```python
xgb_model = XGBClassifier(n_estimators =70, learning_rate = 0.1,verbosity = 1, random_state = RANDOM_STATE)
xgb_model.fit(X_train_fit,y_train_fit, eval_set = [(X_train_eval,y_train_eval)])
```

```python
print(f"Metrics train:\n\tAccuracy score: {accuracy_score(xgb_model.predict(X_train),y_train):.4f}\nMetrics test:\n\tAccuracy score: {accuracy_score(xgb_model.predict(X_val),y_val):.4f}")
```
